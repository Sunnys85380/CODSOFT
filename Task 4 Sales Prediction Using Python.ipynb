```python
# Importing the libraries 
import pandas as pd 
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
```

```python
df = pd.read_csv('advertising.csv') # Data loading
```

# Splitting  and Model Training 

```python
x=df[['TV','Radio','Newspaper']]
y=df['Sales']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test) # Model testing
```

```python
model.score(x_test,y_test) # Accuracy Evaluation
```

```python
# New prediction
'''TV=23
Radio=35
Newspaper=45'''
new_entry=[[23,35,45]]
predicted_sales=model.predict(new_entry)
print('Predicted sales is : ',predicted_sales)
```
