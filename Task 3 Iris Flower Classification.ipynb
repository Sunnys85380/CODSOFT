```python
# Importing the Libraries
import pandas as pd
import matplotlib as plt
%matplotlib inline
import seaborn as sbn
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
df=pd.read_csv('IRIS.csv') # Data Loading
```

# Data Exploration 

```python
print(df)
print("\nall measurements are in centimeters")
```

```python
df.isnull().sum()
```

```python
df.describe()
```

# Visualization

```python
sbn.pairplot(data=df,hue='species',height=2,aspect=1.5) # Iris-setosa significantly differ to Iris-versicolor and Iris-virginica
```

# Splitting, Trainging and Testing

```python
x=df[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
y=df[['species']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
testing=SVC(kernel='linear')
```

```python
testing.fit(x_train,y_train) # Training 
```

```python
testing.predict(x_test) # Model Testing 
```

```python
accuracy=testing.score(x_test,y_test) # Model Accuracy Evaluation
print('Accuracy of Model is :',accuracy*100,'Percent')
```
